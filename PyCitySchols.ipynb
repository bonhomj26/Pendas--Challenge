```python
import pandas as pd
import numpy as np
```

```python
# Read the CSV files
```

```python
schl_data = pd.read_csv("Resources/schools_complete.csv")
schl_data.head()
```

```python
stud_data = pd.read_csv("Resources/students_complete.csv")
stud_data.head()
```

```python
# Renaming name column to school name
schl_data.rename({"name": "school name"}, axis =1, inplace = True)
schl_data.columns
```

```python
schl_data.head()
```

```python
stud_data.rename({"school":"school name"}, axis =1, inplace = True)
stud_data.columns
```

```python
stud_data.head()
```

```python
rename_schl_data = schl_data
rename_schl_data
```

```python
rename_stud_data = stud_data
rename_stud_data
```

```python
# District Summary
# Total Schools
total_schools = rename_schl_data["school name"].count()
total_schools
```

```python
#Total Students
total_students = rename_stud_data["school name"].count()
total_students
```

```python
# Total Budget
tot_budget = rename_schl_data["budget"].sum()
```

```python
# Average Math Score
average_math_score = rename_stud_data["math_score"].mean()
```

```python
# Average Reading Score
average_reading_score = rename_stud_data["reading_score"].mean()
```

```python
# Percent Passing Math 
# Assuming passing = 75
per_pass_math = rename_stud_data.loc[(stud_data["math_score"] >=75)]
per_passm = per_pass_math["math_score"].count()
per_passm
```

```python
percent_passing_math = (per_passm/total_students)*100
percent_passing_math
```

```python
# Percent Passing Reading
# Assuming a passing grade of 75
per_pass_reading = rename_stud_data.loc[(stud_data["reading_score"]>=75)]
per_passr = per_pass_reading["reading_score"].count()
per_passr
```

```python
# % Pass Reading
percent_passing_reading = (per_passr/total_students)*100
percent_passing_reading
```

```python
# Overall Passing Rate
overl_pass_rate = (percent_passing_math + percent_passing_reading)/2
overl_pass_rate
```

```python
# District Summary
district_summary = {"Total Schools":total_schools,
                   "Total Students": total_students,
                   "Total Budget":tot_budget,
                   "Average Math Score":average_math_score,
                   "Average Reading Score":average_reading_score,
                   "%Passing Math":percent_passing_math,
                   "%Passing Reading":percent_passing_reading,
                   "%Overall Passing":overl_pass_rate
                   }
```

```python
district_summary_result = pd.DataFrame([district_summary])
district_summary_result = district_summary_result[["Total Schools","Total Students", "Total Budget","Average Math Score",
                                                  "Average Reading Score", "%Passing Math", "%Passing Reading", "%Overall Passing"]]
```

```python
district_summary_result
```

```python
#School Summary
del rename_schl_data['School ID']
rename_schl_data
```

```python
ren_schl_data = rename_schl_data
ren_schl_data
```

```python
#Percent Student Budget
ren_schl_data['Percent Student Budget'] =ren_schl_data['budget']/ren_schl_data['size']
ren_schl_data
```

```python
# Average Math Score - Average Reading Score
average_mathsc_readsc = rename_stud_data.groupby(['school name'])['reading_score', 'math_score'].mean().reset_index()
average_mathsc_readsc
                                                
```

```python
# Merge dataframes: avg_mathsc_readsc and ren_schl_data
ren_schl_data = ren_schl_data.merge(average_mathsc_readsc, on='school name', how="outer")
ren_schl_data
```

```python
# Percent Passing Math
# Assumming students pass with a score of 75 or greater

schl_passing_math = rename_stud_data[rename_stud_data['math_score']>= 75]
schl_passing_math
```

```python
# Percent Passing Reading
schl_passing_reading = rename_stud_data[rename_stud_data['reading_score']>=75]
schl_passing_reading
```

```python
# Number Student Pasing Math
nbr_stud_passmath = schl_passing_math.groupby(["school name"])['math_score'].count().reset_index()
nbr_stud_passmath.rename({'math_score':'Number MStudents'}, axis =1, inplace = True)
nbr_stud_passmath
```

```python
# Number Students Pass Reading
nbr_stud_passread = schl_passing_reading.groupby(["school name"])['reading_score'].count().reset_index()
nbr_stud_passread.rename({'reading_score': 'Number RStudents'}, axis =1, inplace = True)
nbr_stud_passread
```

```python
# Merge the two dataframes
nbr_studs_pass = nbr_stud_passmath .merge(nbr_stud_passread, on="school name", how='inner')
nbr_studs_pass
```

```python
# Merging ren_schl_data with nbr_studs_pass
ren_schl_data = ren_schl_data.merge(nbr_studs_pass, on='school name', how='outer')
ren_schl_data
```

```python
# Percent Passing Math
ren_schl_data['Percent Passing Math'] = (ren_schl_data['Number MStudents']/ren_schl_data['size'])*100

# Percent Passing Reading
ren_schl_data['Percent Passing Reading'] = (ren_schl_data['Number RStudents']/ren_schl_data['size'])*100
ren_schl_data
```

```python
# The columns "Number MStudents and Number RStudents" were used to avoid having computational error with float and int
# Now the computation is done, they can be deleted

del ren_schl_data['Number MStudents']
del ren_schl_data['Number RStudents']
ren_schl_data
```

```python
# Percent Overall Passing
ren_schl_data['Percent Overall Passing'] = (ren_schl_data['Percent Passing Math'] + ren_schl_data['Percent Passing Reading'])/2
ren_schl_data
```

```python
# Rename the columns
ren_schl_data.rename({'school name': 'School Name', 'type':'School Type','budget':'Total School Budget','Percent Student Budget':'Per Student Budget',
                      'reading_score': 'Average Reading Score', 'math_score':'Average Math Score', 'Percent Passing Math':'% Passing Math',
                      'Percent Passing Reading':'% Passing Reading','Percent Overall Passing':'% Overall Passing'}, axis =1, inplace = True)
ren_schl_data
```

```python
school_summary = ren_schl_data
school_summary
```

```python
# Top Peforming Schools
top_perfmg_schls = school_summary.sort_values(by = ['% Overall Passing'], ascending=False)
top_perfmg_schls
```

```python
# Bottom Peforming Schools
bot_perfmg_schls = school_summary.sort_values(by = ['% Overall Passing'], ascending=False)
bot_perfmg_schls
```
